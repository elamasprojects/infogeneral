# Operaciones

### Operadores de comparación **< > <= >= == !=**

Los operadores de comparación permiten comparar dos valores, y retornan **siempre** un valor de tipo de dato booleano ```bool```.

````python
a = 1
b = 2
````

````python
a==b
````

````python
b<a
````

````python
a<=b
````

````python
a!=b
````

````python
5 == 5.0
````

La forma en que se comparan los valores depende del tipo de valor. Por ejemplo, los números se comparan utilizando su escala numérica, mientras que las cadenas de caracteres ```str``` se comparan carácter por carácter.

Las reglas para la  comparación de cadenas son:

* Una cadena ```str``` es considerada menor que otra al evaluar los caracteres uno por uno, utilizando la [tabla ascii](http://1.bp.blogspot.com/-gTya0k5EPtY/UlsYUkVM-HI/AAAAAAAAAIg/YeMEivnolAY/s1600/IMAGEN13.jpg) para el valor de cada caracter.
* Si el primer caracter en ambas cadenas es el mismo, entonces se pasa a evaluar el siguiente en ambas cadenas.
* Si todo los caracteres evaluados son iguales en ambas cadenas, pero una cadena es más larga que la otra, se considera que la cadena más corta es menor.

### Tabla ascii
Es un código de caracteres que nos permite comunicarnos con el computador a través de nuestro lenguaje nativo, en nuestro caso nos permite escribir las instrucciones usando el alfabeto latino.

## Características
* No es un lenguaje de programación.
* Es un patrón estandard que permite representar 127 caracteres o símbolos.
* Cada símbolo del código Ascii está asociado a un número que indica su posición dentro de la tabla de caracteres.
* Los primeros símbolos desde el 0 al 31 corresponden a caracteres de control no imprimibles, por ejemplo cambio de linea, tabulador, retroceder, etc
* Los símbolos correspondientes entre las posiciones 65 y 90 representan las letras mayúsculas del alfabeto.
* Los símbolos correspondientes entre las posiciones 48 y 57 representan los caracteres numéricos entre 0 y 9.
* Para transformar los símbolos de las letras mayúsculas a minúsculas se le debe sumar 32 y viceversa.

## Operadores lógicos and, or y not (unario)

Los operadores lógicos siempre devuelven un resultado ```bool``` como resultado.

````python
verdad = True
mentira = False
````

````python
verdad ==True
````

````python
verdad and True     # and devuelve True si ambos operando son True
````

````python
mentira and True    # mentira es false
````

````python
mentira or verdad  # El or devuelve true si alguno de los
                   # dos operandos es verdadero
````

````python
not verdad          # not invierte el valor del bool.
                    # False pasa a ser True y True  pasa a ser false
````

````python
not not True        # Aunque no tiene sentido, se puede usar mas de una vez
````

````python
not  True and False  # También se puede aplicar sobre el resultado de otra
                       # operación, Los paréntesis, igual que en matemática
                       # me permiten decirle al intérprete que tiene
                       # que evaluar primero
````

Si en una expresión hay más de un operador del mismo nivel de precedencia, tiene prioridad el que está más a la izquierda en la expresión, con la excepción de **.


Python hace una diferencia entre convertir a ```bool``` (usando la función ```bool()```) e interpretar como bool para las operaciones lógicas.

````python
bool(5) and bool(6)     ## Se convierte a bool para ser analizados
````

````python
4 and 6     ## Cuando se interpreta dos valores,
            # el and devuelve el de la derecha
````

````python
4 and 0
````

````python
5 or 6     ## Cuando se interpreta dos valores, el or devuelve
           #  el de la izquierda si es interpretado como verdadero
````

````python
0 or 7      ## Cuando se interpreta dos valores, el or devuelve el de la
            # derecha si el de la izquiera es 0, ya que este se interpreta como falso.
````

### Evaluación mínima

El intérprete al encontrar una operación booleana, va a ir evaluándola y apenas ya sepa el resultado, deja de evaluar y el resto de la operación es ignorada.

````python
False and True
````

````python
variable_no_definida    ## Si trato de usar una variable no definida,
                        # el intérprete me tira un error
````

````python
a = False
a and variable_no_definida   ## Como nunca llega a evaluar el valor de la
                             # variable no definida, acá no me tira un error
````

````python
a = True
a or variable_no_definida
````

````python
a = 0
1 / a   ## La división por cero no se permite
````

````python
a = 0
bool(a) and (a/a)==1  ## Cómo a es interpretado como falso, 0/0 no se evalúa.
````

````python
a = 5
bool(a) and (a/a)==1  ## Cómo a es interpretado como Verdadero,
                      # se evalúa a/a, que da uno y la comparación
                      # con 1 es verdadera
````

## Indices y slices en str

Las cadenas pueden ser accedidas caracter por caracter usando un índice. Los índice se especifican por medio de corchetes `[i]`, donde `i` representa la posición del caracter dentro del str.

El índice 0 representa el primer caracter, y a partir de cero se va incrementando. [1] es el segundo caracter y desde ahí hasta el final.Dado que comienza con 0, la última posición será el valor de la longitud de la cadena menos uno.

Un índice negativo representa la posición contando desde el último caracter hacia adelante, comenzando en -1.

````python
s = 'abcdefghijkmnopq'  ## Defino un str
s[2]                    ## Caracter en el índice 2, que es la terecera posición
````

````python
s[-1]                   ## -1 representa el último caracter
````

Si en el corchete agrego más de un número separado por `:` python extrae un rango de caracteres. El formato es `[start:stop:step]` donde:
* `start`: posición del primer caracter a sacar. Por default: el primero disponible (depende de step)
* `stop`: siguiente posición al último a sacar. Por default: el siguiente al último disponible (depende de step)
* `step`: el salto o incremento entre los índice seleccionados. Por default: 1

````python
s = 'abcdefghijkmnopq'
print(s[0:5])               ## Comenzando en cero (el primero) y
                            #  5 es la posición siguiente al último que quiero
print(s[5])                 # Caracter en la posición 5 para tener de referencia
````

````python
s[:]                       ## Usando los valores default, no tiene cambios
````

````python
s[5:]                       ## Arrancando en la posición 5,
                            # que ahora sí está incluida.
````

````python
s = 'abcdefghijkmnopq'
s[-1]                       ## Último caracter
````

````python
s[-6:]            ## Los últimos 6 caracteres ( start=-6, stop: último , step=1 )
````

````python
s[-6::-1]       ## Los primeros 6 caracteres en orden inverso ( start=-6, stop: último , step=-1 )
````

````python
s[-1::-1]                   ## Toda la cadena de atrás hacia adelante
````

````python
print(s[::1])        ## el default de primero y último depende de step. step=1
print(s[::-1])       ## step = -1
````

````python
s[100]      # Si uso un valor de índice inválido obtengo un error de IndexError
````
